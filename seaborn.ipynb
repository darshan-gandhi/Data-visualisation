```python
titanic=sns.load_dataset('titanic')
sns.set_style("ticks")
sns.set_context("talk")
sns.barplot("class","survived", "sex", data = titanic, palette="pastel")

plt.show()
```

```python
titanic=sns.load_dataset('titanic')
sns.set_style("ticks")
sns.set_context("talk")
ax=sns.catplot("class","survived", "sex", kind="bar", data = titanic, palette="pastel" , height=8, aspect=2)
ax.fig.suptitle("Titanic Survivor by class")

plt.show()
```
