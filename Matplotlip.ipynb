```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import pandas as pd
```

```python
student_names = ['Jamal', 'Kamal', 'Rony', 'Jony', 'Sumon']
math_result = dataset['math'][:5]


plt.bar(student_names, math_result)
```

```python
fig, ax = plt.subplots(1,2, figsize= (13,5))
ax[0].bar(student_names, math_result)
ax[0].legend('M')
plt.title('Final result')

ax[1].bar(student_names, math_result, color = 'orange')
ax[1].legend("M")
plt.title('final exam', fontsize = 18 )
plt.xticks(fontsize= 13)
plt.yticks(fontsize =13)
plt.show()
```

```python
N= 4
dataset_mean = dataset.mean()
dataset_std = dataset.std()

print ('mean' , dataset_mean)
print ('std' , dataset_std)

ind = np.arange(N)
width = 0.35

p1 = plt.bar(ind, dataset_mean, width, yerr= dataset_std)

plt.ylabel('Scores')
plt.title('Dataset Visualization')
plt.xticks(ind, ("Math", "Bangla", "English", "Result"))
plt.yticks(np.arange(0,81,10))

plt.show()
```

```python
N = 50 #number of random values flag

x = np.random.rand(N)
y = np.random.rand(N)

colors = np.random.rand(50) #random colors

area = ( 30 * np.random.rand(N))**2 #scater random area

plt.scatter(x,y, label = "Scatter plot", s= area, c = colors, alpha = 0.5) # alpha for scatter opacit y, alpha = 1 means solid solor
plt.legend()

plt.xlabel= ("X- axias")
plt.ylabel = ("Y-axias")
plt.title("Scatter plot example")

plt.show()
```
